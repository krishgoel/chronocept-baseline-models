```python
import torch
from utils import DataLoader, evaluate_model
from models import BiLSTMModel
from matplotlib import pyplot as plt
```

```python
import logging
logging.basicConfig(level=logging.INFO)
```

```python
dl = DataLoader(
    benchmark="benchmark_1",
    split=None,
    embedding="bert_sequential", 
    max_len=16,
    include_axes=True,
    shuffle_axes=False,  
    normalization="zscore"
)
```

```python
data = dl.preprocess()
X_train, y_train = data["train"]
X_valid, y_valid = data["validation"]
X_test, y_test = data["test"]
```

```python
save_path = "baseline_weights/benchmark_1/bilstm.pth"
input_dim = 768
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")
```

```python
grid_params = {
    "hidden_dim": [64, 128, 256],
    "lr": [0.01, 0.001, 0.0001]
}
epochs = 100
```

```python
dummy_model = BiLSTMModel(input_dim)
best_model, best_params, train_losses, valid_losses = dummy_model.grid_search(
    train_data=(X_train.cpu().numpy(), y_train),
    valid_data=(X_valid.cpu().numpy(), y_valid),
    grid_params=grid_params,
    epochs=epochs,
    device=device
)

print("Best Hyperparameters:", best_params)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label="Train Loss")
plt.plot(valid_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.title(f"LSTM Performance {best_params}")
plt.legend()
plt.savefig("plots/benchmark_1/lstm_grid_search.pdf")
plt.show()
```

```python
best_model.eval()
with torch.no_grad():
    test_inputs = torch.tensor(X_test.cpu().numpy(), dtype=torch.float32).to(device)
    test_preds = best_model(test_inputs).cpu().numpy()
```

```python
mse, mae, r2, nll, crps = evaluate_model(y_test, test_preds)
print(f"MSE: {mse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}, NLL: {nll:.4f}, CRPS: {crps:.4f}")
```

```python
best_model.save(save_path)
```
